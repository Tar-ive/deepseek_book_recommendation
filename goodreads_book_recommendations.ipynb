```python
file_name_type_mapping = dict(zip(file_names['name'].values, file_names['type'].values))
file_name_url_mapping = {}

for fname in file_name_type_mapping:
    ftype = file_name_type_mapping[fname]
    if ftype == "complete":
        url = 'https://mcauleylab.ucsd.edu/public_datasets/gdrive/goodreads/'+fname
        file_name_url_mapping[fname] = url
    elif ftype == "byGenre":
        url = 'https://mcauleylab.ucsd.edu/public_datasets/gdrive/goodreads/byGenre/'+fname
        file_name_url_mapping[fname] = url
```

```python
def download_by_name(fname, local_filename):
    if fname in file_name_url_mapping:
        url = file_name_url_mapping[fname]
        with requests.get(url, stream=True) as r:
            r.raise_for_status()
            with open(local_filename, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
        print('Dataset', fname, 'has been downloaded!')
    else:
        print('Dataset', fname, 'can not be found!')
```

```python
def parse_fields(line):
  data = json.loads(line)
  return{
      'book_id': data['book_id'],
      'title': data['title_without_series'],
      'authors': data['authors'],
      'ratings': data['ratings_count'],
      'url': data['url'],
      'cover_image': data['image_url']
  }
```

```python
book_titles = []
with open('genre/goodreads_books.json', 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        fields = parse_fields(line)
        try:
          ratings = int(fields['ratings'])
        except ValueError:
          continue
        if ratings > 15:
          book_titles.append(fields)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

tfidf = vectorizer.fit_transform(titles['mod_title'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import re

def make_clickable(val):
  return '<a target="_blank" href="{}">Goodreads</a>'.format(val)

def show_image(val):
  return '<a href="{}"><img src="{}" width=50></img></a>'.format(val, val)

def search(query, vectorizer):
  processed = re.sub('[^a-zA-Z0-9 ]', '', query.lower())
  query_vec = vectorizer.transform([processed])
  similarity= cosine_similarity(query_vec, tfidf).flatten()
  indices = np.argpartition(similarity, -10)[-10:]
  results = titles.iloc[indices]
  results = results.sort_values('ratings', ascending=False)
  return results.head(5).style.format({'url': make_clickable, 'cover_image': show_image})
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import re

def verify_book_id(title, book_id, titles_df):
    """Verify a book ID matches the title"""
    book_info = titles_df[titles_df['book_id'] == book_id]
    if not book_info.empty:
        book = book_info.iloc[0]
        print(f"\nVerifying: {title}")
        print(f"Book ID: {book_id}")
        print(f"Found title: {book['title']}")
        print(f"Ratings: {book['ratings']}")
        print(f"URL: {book['url']}")
        return True
    return False

def search_and_verify(query, vectorizer, tfidf_matrix, titles_df):
    """Search for a book and return top matches"""
    processed = re.sub('[^a-zA-Z0-9 ]', '', query.lower())
    query_vec = vectorizer.transform([processed])
    similarity = cosine_similarity(query_vec, tfidf_matrix).flatten()
    indices = np.argpartition(similarity, -5)[-5:]
    results = titles_df.iloc[indices]
    results = results.sort_values('ratings', ascending=False)
    return results.head(5)

def create_liked_books_csv(vectorizer, tfidf_matrix, titles_df):
    """Create liked_books.csv with verified book IDs"""

    # Initial book data with known IDs
    books_data = [
        ('White Nights', '1772910', 5),
        ('Pride and Prejudice', '14918', 5),
        ('The Metamorphosis', '7718', 5),
        ('Harry Potter and the Prisoner of Azkaban', '464164', 5),
        ('Crime and Punishment', '160296', 4),
        ('The Master and Margarita', '117833', 4),
        ('The Secret History', '13054604', 4),
        ('Never Let Me Go', '102927', 4),
        ('One Hundred Years of Solitude', '833213', 4),
        ('White Nights', '1772910', 5),
        ('The Master and Margarita', '117833', 4),
        ('The Metamorphosis', '7718', 5),
        ('Crime and Punishment', '160296', 3),
        ('Harry Potter and the Prisoner of Azkaban', '464164', 5),
        ('Harry Potter and the Goblet of Fire', '28754622', 4),
        ('The Name of the Wind', '186074', 3),
        ('Pride and Prejudice', '14918', 5),
        ('Emma', '894048', 4),
        ('Sense and Sensibility', '8130026', 4)
    ]

    # Verify each book ID and create final data
    verified_data = []
    for title, book_id, rating in books_data:
        if verify_book_id(title, book_id, titles_df):
            verified_data.append({
                'user_id': 0,
                'book_id': book_id,
                'rating': rating,
                'title': title
            })
        else:
            print(f"\nWARNING: Could not verify book ID {book_id} for {title}")
            print("Searching for alternatives:")
            alternatives = search_and_verify(title, vectorizer, tfidf_matrix, titles_df)
            print(alternatives[['book_id', 'title', 'ratings']])

    # Create DataFrame and save to CSV
    df = pd.DataFrame(verified_data)
    df.to_csv('liked_books.csv', index=False)
    return df

# Usage example:

# Your existing code for loading and processing titles should be run first
liked_books_df = create_liked_books_csv(vectorizer, tfidf, titles)
print("\nFirst few rows of created dataset:")
print(liked_books_df.head())
```

# using deepseek to rerank the recommendations

```python
from google.colab import userdata
api_key = userdata.get('DEEPSEEK_API_KEY')
```

```python
from openai import OpenAI
import pandas as pd
import numpy as np
import json
import time

def create_prompt(book):
    """Create prompt for evaluating a single book"""
    prompt = f"""Given a reader who loves these specific books:

1. White Nights by Dostoevsky - psychological realism, introspective narrative
2. Pride and Prejudice by Austen - social commentary, complex relationships
3. The Metamorphosis by Kafka - surrealism, existential themes
4. Harry Potter and the Prisoner of Azkaban by Rowling - fantasy, coming of age
5. The Picture of Dorian Gray by Wilde - aestheticism, moral decay
6. Notes from Underground by Dostoevsky - existentialism, psychological depth

Evaluate this book for the reader:
Title: {book['title']}
Author(s): {book['authors']}

Rate from 0-100 how well this book matches their taste. Consider:
1. Writing style
2. Themes
3. Psychological depth
4. Literary complexity
5. Character development

Format response as JSON:
{{"score": <number>, "reason": "<brief explanation>"}}"""

    return prompt

def process_book(book, client):
    """Process a single book with the LLM"""
    try:
        response = client.chat.completions.create(
            model="deepseek-chat",
            messages=[
                {"role": "system", "content": "You are a literary expert. Response must be valid JSON."},
                {"role": "user", "content": create_prompt(book)}
            ],
            temperature=0.7,
            max_tokens=150
        )

        result = json.loads(response.choices[0].message.content)
        return {
            'book_id': book['book_id'],
            'title': book['title'],
            'authors': book['authors'],
            'original_score': book['score'],
            'count': book['count'],
            'mean_rating': book['mean'],
            'llm_score': result['score'],
            'explanation': result['reason'],
            'url': book['url'],
            'cover_image': book['cover_image']
        }
    except Exception as e:
        print(f"Error processing {book['title']}: {str(e)}")
        return None

def rerank_recommendations(top_recs_df):
    """Rerank the filtered recommendations using LLM"""
    try:
        client = OpenAI(
            api_key=userdata.get('DEEPSEEK_API_KEY'),
            base_url="https://api.deepseek.com"
        )

        print(f"Processing {len(top_recs_df)} recommendations...")
        results = []

        for _, book in top_recs_df.iterrows():
            result = process_book(book, client)
            if result:
                results.append(result)
            time.sleep(1)

        if not results:
            return None

        # Create final dataframe
        final_df = pd.DataFrame(results)

        # Normalize scores
        final_df['normalized_original'] = (final_df['original_score'] - final_df['original_score'].min()) / \
                                        (final_df['original_score'].max() - final_df['original_score'].min())
        final_df['normalized_llm'] = final_df['llm_score'] / 100

        # Combine scores (40% original, 60% LLM)
        final_df['final_score'] = (0.4 * final_df['normalized_original'] +
                                 0.6 * final_df['normalized_llm'])

        return final_df.sort_values('final_score', ascending=False)

    except Exception as e:
        print(f"Error in reranking: {str(e)}")
        return None

def display_recommendations(df, n=10):
    """Display top N recommendations with formatting"""
    if df is None or len(df) == 0:
        print("No recommendations available")
        return

    top_n = df.head(n)

    # Display formatted table
    display(top_n.style.format({
        'url': make_clickable,
        'cover_image': show_image,
        'llm_score': '{:.1f}',
        'final_score': '{:.3f}'
    }))

    # Display detailed explanations
    print("\nDetailed Recommendations:")
    for _, book in top_n.iterrows():
        print(f"\n{book['title']}")
        print(f"Original Score: {book['original_score']:.3f}")
        print(f"LLM Score: {book['llm_score']:.1f}")
        print(f"Final Score: {book['final_score']:.3f}")
        print(f"Reason: {book['explanation']}")

# Example usage:

# Make sure top_recs is your filtered DataFrame from collaborative filtering
# It should already have count > 2 and mean > 4

# Rerank recommendations
reranked_books = rerank_recommendations(top_recs)

# Display results
if reranked_books is not None:
    display_recommendations(reranked_books, n=10)
```
